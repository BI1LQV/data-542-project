```python
import pandas as pd
# read our dataset
df = pd.read_json('../data/20230817_130502_issue_sharings.json')
df = pd.json_normalize(df['Sources'])
ChatgptSharing =  pd.json_normalize(df['ChatgptSharing'])

# get some insight about languages
repo_lang_counts = df['RepoLanguage'].value_counts()
repo_lang_counts
```

```python
ChatgptSharing = pd.json_normalize(df['ChatgptSharing'].explode())

conversations = pd.json_normalize(ChatgptSharing['Conversations'].explode())

# parse all code blocks and get the actual count
code_blocks = pd.json_normalize(conversations['ListOfCode'].explode())
code_blocks = code_blocks.dropna(how='all')

code_blocks_type_counts = code_blocks['Type'].value_counts()
code_blocks_type_counts
```

```python
import pandas as pd
import altair as alt

code_blocks_type_counts = code_blocks_type_counts.sort_values(ascending=False)

# <60 to another
threshold = 60
filtered_langs = code_blocks_type_counts[code_blocks_type_counts >= threshold]
other_count = code_blocks_type_counts[code_blocks_type_counts < threshold].sum()

lang_data = pd.DataFrame({
    'Language': list(filtered_langs.index) + (["Other"] if other_count > 0 else []),
    'Count': list(filtered_langs.values) + ([other_count] if other_count > 0 else [])
})

# sort
lang_data = lang_data.sort_values('Count', ascending=False)

# get pie chart
pie_chart = alt.Chart(lang_data).mark_arc().encode(
    theta='Count:Q',
    color=alt.Color('Language:N'),
    tooltip=['Language', 'Count']
).properties(
    width=400,
    height=400,
    title='Code Block Language Distribution'
)

pie_chart
```

```python
python_code_blocks = code_blocks[code_blocks['Type']=='python']['Content']

import os
import shutil

# get all python codes and write them into python_codes

if os.path.exists('./python_codes'):
    shutil.rmtree('./python_codes')

os.makedirs('./python_codes')

for idx, code in python_code_blocks.items():
    filename = f'./python_codes/{idx}.py'
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(code)
```

```python
javascript_code_blocks = code_blocks[code_blocks['Type']=='javascript']['Content']


# get all js codes and write them
if os.path.exists('./js_codes'):
    shutil.rmtree('./js_codes')

os.makedirs('./js_codes')

for idx, code in javascript_code_blocks.items():
    filename = f'./js_codes/{idx}.js'
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(code)
```
